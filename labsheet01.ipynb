```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[4,5,6,7]
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("line graph")
plt.plot(x,y)
plt.show()
```

```python
y= [10,20,30,60]
mylabels = ["Apples","Bananas","Cherries","Dates"]
mycolors = ["black", "hotpink","b","#4CAF50"]
myexplode=[0.5, 0 , 0 ,0.5]
plt.pie(y, labels = mylabels,colors = mycolors,explode = myexplode)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array(["A","B","C","D"])
y=np.array([3,6,1,10])
plt.bar(x,y,color="r")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 6, 1, 10])
mycolor = np.array(["r", "b", "c", "k"])
plt.barh(x,y,color=mycolor)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,5,6,7]
mycolor=["r","y","b","k","c"]
sizes=[100,200,150,300,400]
plt.title("scatter",fontsize=20)
plt.xlabel("x axis",fontsize=20)
plt.ylabel("y axis",fontsize=20)
plt.scatter(x,y,c=mycolor,s=sizes,alpha=0.5,marker="*",edgecolor="black")
plt.show()
```
